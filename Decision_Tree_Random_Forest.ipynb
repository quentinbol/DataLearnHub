```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("./dataset/dataset/DecisionTree_PastHires.csv")
```

```python
df
```

```python
df.keys()
```

```python
df['Level of Education'].value_counts()
```

```python
d1={'Y':1,'N':0}
d2={'BS':1,'MS':2,'PhD':3}
```

```python
df['Employed?']=df['Employed?'].map(d1)
df['Top-tier school']=df['Top-tier school'].map(d1)
df['Interned']=df['Interned'].map(d1)
```

```python
df['Level of Education']=df['Level of Education'].map(d2)
df['Hired']=df['Hired'].map(d1)
```

```python
df
```

```python
model=DecisionTreeClassifier(criterion='gini',random_state=42)
model2=RandomForestClassifier(n_estimators=100, random_state=2)
```

```python
X=df.iloc[:, :-1].values
print(X)
Y=df.iloc[:, -1].values
print(Y)
X_train,X_test,Y_train, Y_test=train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
model.fit(X_train,Y_train)
model2.fit(X_train,Y_train)
y_pred=model.predict(X_test)
y_pred2=model2.predict(X_test)
```

```python
print("precision", metrics.accuracy_score(Y_test, y_pred))
```

```python
print("precision", metrics.accuracy_score(Y_test, y_pred2))
```
