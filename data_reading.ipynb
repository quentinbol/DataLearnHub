```python
prix_ms=[39,50,57,69,94,100]
prix_apple=[93,112,104,144,170,172]
annee=[2015,2016,2017,2018,2019,2020]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#visualisation dans 1 plan
plt.scatter(annee,prix_ms, label='Prix des ordinateurs Windows', c='g')
plt.scatter(annee,prix_apple, label='Prix des ordinateurs Apple', c='r')
plt.legend()
plt.xlabel('Année')
plt.ylabel('Prix des Ordinateurs')
plt.title('Evolution des Prix des Ordinateurs durant les 5 Dernières Années')
```

```python
#Visualisation dans 2 plans
plt.subplot(2,1,1)
plt.scatter(annee,prix_ms, label='Prix des ordinateurs Windows', c='lime')
plt.subplot(2,1,2)
plt.scatter(annee,prix_apple, label='Prix des ordinateurs Apple', c='r')
plt.legend()
```

```python
#Visualisé les données en diagramme de bande
index_annee=np.arange(len(annee))
size=0.25
plt.bar(index_annee,prix_ms, label='Prix des ordinateurs Windows', width=0.25)
plt.bar(index_annee+size,prix_apple, label='Prix des ordinateurs Apple', width=0.25)
plt.legend()
plt.xlabel('Année')
plt.ylabel('Prix des Ordinateurs')
plt.title('Evolution des Prix des Ordinateurs durant les 5 Dernières Années')
```
